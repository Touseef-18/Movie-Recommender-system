```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("C:\\Users\\LENOVO\\Downloads\\tmdb_5000_movies.csv\\tmdb_5000_movies.csv")
movies.head(1)
```

```python
credits = pd.read_csv("C:\\Users\\LENOVO\\Downloads\\tmdb_5000_credits.csv\\tmdb_5000_credits.csv")
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
x= movies.iloc[0][3]
```

```python
import ast
y = ast.literal_eval(x)
```

```python
y[0]['name']
```

```python
def genre(obj):
    for i in ast.literal_eval(obj):
        return i
```

```python
def genres(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
        
```

```python
movies['genres'] = movies['genres'].apply(genres)
movies.head()
```

```python
movies.iloc[0][4]
```

```python
def keywords(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['keywords'] = movies['keywords'].apply(keywords)
movies.head()
```

```python
def cast(obj):
    l=[]
    count = 0 
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count = count +1
        else:
                break
    return l
        
```

```python
movies['cast'] = movies['cast'].apply(cast)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
x=movies.iloc[0][5]
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector
```

```python
list(enumerate(vector))
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```
